```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#method part 1, force vs current
#plotting F against I squared
Fpt1 = (10**-5)*np.array([9.81, 14.72, 19.62, 24.53, 29.43, 34.34])
ErrorFpt1 = (10**-5)*np.array([0.005, 0.005, 0.005, 0.005, 0.005, 0.005])

IsquaredPt1 = np.array([9, 17.6, 34.8, 38.4, 42.2, 57.7])
ErrorIsquaredPt1 = np.array([0.1, 0.1, 0.2, 0.2, 0.2, 0.2])
```

```python
x = Fpt1
y = IsquaredPt1
xerr = ErrorFpt1
yerr = ErrorIsquaredPt1


plt.xlabel('Force')
plt.ylabel('Current squared')
plt.errorbar(x, y, yerr, xerr, fmt='none', ecolor='r')
def fitfunc(x,m,c):
    return m*x+c
pars, cov = curve_fit(fitfunc,x,y)
plt.plot(x,fitfunc(x,*pars),'g-')
print(pars)#first element is slope
print(cov)#top left element is square of error on slope
```

```python
#method pt 2, force vs separation
#need to plot F against 1/r

Fpt2 = 10**(-4)*np.array([10, 9.82, 9.45, 8.71, 7.78, 6.67, 5.93, 5.56, 5.37,5.19])
Fpt2err = 0.005*10**(-4)
rPt2 = 10**(-3)*np.array([4.1, 5.1, 6.1, 7.1, 8.1, 9.1, 10.1, 11.1, 12.1, 13.1])
rpt2err = 0.1*10**(-3)
```

```python
x = Fpt2
y = rPt2
xerr = Fpt2err
yerr = rpt2err


plt.xlabel('Force')
plt.ylabel('1/r')
plt.errorbar(x, y, yerr, xerr, fmt='none', ecolor='r')
def fitfunc(x,m,c):
    return m*x+c
pars, cov = curve_fit(fitfunc,x,y)
plt.plot(x,fitfunc(x,*pars),'g-')
print(pars)#first element is slope
print(cov)#top left element is square of error on slope
```
